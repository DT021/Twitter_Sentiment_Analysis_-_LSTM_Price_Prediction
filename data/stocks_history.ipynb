```python
import alpaca_trade_api as tradeapi
import config
from datetime import datetime, timedelta, date
import pandas as pd
```

```python
api = tradeapi.REST(
    config.API_KEY,
    config.SECRET_KEY,
    api_version='v2'
)
```

```python
timeframe = '1D' #1Min, 5Min, 15Min, 1D
tickers = ['MSFT', 'AAPL', 'BRK.B']
limit = '1000' #between 1 and 1000. default is 100.
end_date = date.today()
start_date = end_date - timedelta(365*5)
```

```python
stock_history_df = api.get_barset(
    symbols=tickers,
    timeframe=timeframe,
    limit=limit,
    start=start_date,
    end=None,
    after=None,
    until=None
).df
```

```python
stock_history_df.dropna(inplace=True)
```
